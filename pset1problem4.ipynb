```python
import math
from numpy import random

def serial_dilution(f_0, N, s, mu):
    population = [0] * N
    f = [f_0]
    delta_t = 1
    for t in range(0, 2000):
        # dilute
        dilute = random.poisson(N * ( 1 - mu * delta_t) * (1 - f[t]) / (1 - f[t] + f[t] * math.exp(s * delta_t)), 100)
        population = [0] * (dilute) 
        # undilute population
        population *= dilute
        # calculate frequency
        if (s!= 0):
            f.append((f_0 * math.exp(s * t * delta_t))/(1 - f_0 + f_0 * math.exp(s * (t + 1) * delta_t)))
        else:
            f.append((f[t] + math.sqrt(f[t]/N))/(1 + math.sqrt(f[t]/N) + math.sqrt(1-f[t]/N)))
        
    return f[:-1]
```

```python
import matplotlib.pyplot as plt

# part a                                               
f_0 = 0.5
N = [10**2, 10**3, 10**6]
s = [0, 10**(-2), 10**(-3)]                                
mu = 0

f = serial_dilution(f_0, 10**2, 0, mu)
plt.plot(list(range(0, 2000)), f)
plt.title("s = 0, N = 10^2, f(0) = 0.5, mu = 0")
plt.ylabel("mutant frequency")
plt.xlabel("number of transfers")
plt.show()

f = serial_dilution(f_0, 10**2, 10**(-2), mu)
plt.plot(list(range(0, 2000)), f)
plt.title("s = 10^-2, N = 10^2, f(0) = 0.5, mu = 0")
plt.ylabel("mutant frequency")
plt.xlabel("number of transfers")
plt.show()
           
                                               
```
